```python
import random


def gen_matrix(rows, cols,
    distribution="uniform",
    value_range=(0.0, 1.0),
    mean=0.0, stddev=1.0, 
    decimals=2):

    matrix = []

    for i in range(rows):
        row = []
        for j in range(cols):

            if distribution == "uniform":
                value = random.uniform(*value_range)

            else:
                value = random.gauss(mean, stddev)

            row.append(round(value, decimals))
        matrix.append(row)
    return matrix
```

```python
matrix = gen_matrix(
    rows=256,
    cols=256,
    value_range=[-1.0, 1.0],
    decimals=16)

print(matrix)
```
